```python
#Connecting to webcam
vcap = cv2.VideoCapture(0)
while vcap.isOpened(): 
    ret , frame = vcap.read()
    cv2.imshow("Image Collection", frame)#return image back to screen
     #waits 1 milisecond and checks what key to press to break vcap - input on keybaord   
    if cv2.waitKey(1) & 0XFF == ord("q"):
        break
vcap.release()
cv2.destroyAllWindows()
```
